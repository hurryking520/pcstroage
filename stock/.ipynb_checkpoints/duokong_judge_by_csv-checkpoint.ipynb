```python
import pandas as pd
import numpy as np
```

```python
#['open','close','high','low','rise','exchange','ma5','ma10','ma20','ma30','ma60']
test_data = np.loadtxt('603127.csv',delimiter=',')
```

```python
print(test_data)
```

```python
def sigmond(x):
    return 1/(1+np.exp(-x))
```

```python
def nor(test_data):
    x = test_data-np.mean(test_data,axis=0)
    print(x)
    x = x / np.std(test_data,axis=0)
    print(x)
    return x
```

```python
#################读取w1 与w2##################################
w1 = np.loadtxt("init_w1.txt",delimiter=',')
w2 = np.loadtxt("init_w2.txt",delimiter=',')
w2 = w2.reshape(-1,1)
#################根据文件内容，初始化w1与w2#####################

#print(w2[0:10,:])
print(w1[0:10,0:5])
print(w2[0:10])
```

```python
x = nor(test_data)

x = x.reshape(1,-1)
x = np.insert(x,0,1,axis=1)
print(x.shape)
hid_input = np.matmul(x,w1)
hid_output = np.tanh(hid_input)
hid_output = np.insert(hid_output,0,1,axis=1)
hpfs_input = np.matmul(hid_output,w2)
hpfs_output = sigmond(hpfs_input)
print(hpfs_output)
```
